```python
import numpy as np
import glob
import os
import scipy.io
import matplotlib.pyplot as plt

REL_PATH = '/home4/shubham/MTML_Pth/datasets/SUNRGBD/'
```

```python
def load_dataset():
    """
    returns a dictionary of train, test and validation images and their corresponding segmentation maps
    """
    images = []
    seg = []

    for filename in glob.iglob(REL_PATH + '**/image/*.jpg', recursive=True):
        images.append(filename)
        path = '/'.join(filename.split('/')[:-2])+"/seg.mat"
        seg.append(path)
    
    return images, seg
```

```python
images, labels = load_dataset()
```

```python
# example of segmentation map array

mat = scipy.io.loadmat(labels[33])
print("Image : ",mat['seglabel'])
print("Max pixel value : ",np.max(mat['seglabel']))
plt.imshow(mat['seglabel'])
```
